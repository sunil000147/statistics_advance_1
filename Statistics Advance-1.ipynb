1.It must be non-negative for all values of the random variable.

Q2. What are the types of Probability distribution?



1.Discrete probability distribution: A probability distribution where the random variable can take on only a finite or countably infinite number of values. Examples include the binomial distribution, Poisson distribution, and geometric distribution.

2.Continuous probability distribution: A probability distribution where the random variable can take on any value in a continuous range. Examples include the normal distribution, exponential distribution, and beta distribution.

3.Bernoulli distribution: A probability distribution where the random variable can take on only two values, usually 0 or 1, representing failure or success, respectively.

4.Binomial distribution: A discrete probability distribution where the random variable represents the number of successes in a fixed number of independent trials, with a constant probability of success in each trial.

5.Poisson distribution: A discrete probability distribution that models the number of occurrences of a rare event in a fixed amount of time or space.

6.Normal distribution: A continuous probability distribution that is widely used to model real-world phenomena. It has a bell-shaped curve and is characterized by its mean and standard deviation.

7.Exponential distribution: A continuous probability distribution that models the time between events in a Poisson process.

8.Gamma distribution: A continuous probability distribution that is a generalization of the exponential distribution and can be used to model a variety of phenomena, such as waiting times, income distribution, and insurance claims.

9.Beta distribution: A continuous probability distribution that is often used to model the probability of success in a Bernoulli trial, where the probability of success is itself a random variable.

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import numpy as np
import math

def normal_pdf(x, mu, sigma):
    """
    Calculates the probability density function (PDF) of a normal distribution
    at a given point x, with mean mu and standard deviation sigma.
    """
    pdf = (1 / (sigma * math.sqrt(2 * math.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))
    return pdf
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.


Ans.Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has the same probability of success.

The properties of a binomial distribution are:

The experiment consists of a fixed number of trials, denoted by n.
Each trial has only two possible outcomes, usually called success and failure.
The probability of success is denoted by p and remains constant throughout the experiment.
The trials are independent of each other, meaning that the outcome of one trial does not affect the outcome of any other trial.
The random variable X represents the number of successes in n trials.
Examples of events where binomial distribution can be applied include:

A manufacturing company wants to test the quality of its products. They take a sample of 100 products and test each one for defects. They record the number of defective products in the sample. This can be modeled as a binomial distribution, where the number of defective products is the random variable X, the number of trials is n=100, and the probability of a defective product is p.

A political pollster wants to estimate the proportion of voters in a particular city who support a certain candidate. They randomly select 500 voters and ask each one whether they support the candidate. The number of voters who support the candidate can be modeled as a binomial distribution, where the number of trials is n=500, and the probability of a voter supporting the candidate is p.

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.





```python
import math

def poisson_cdf(mean, point):
    cdf = 0
    for i in range(point + 1):
        cdf += math.exp(-mean) * (mean ** i) / math.factorial(i)
    return cdf
```

Q7. How Binomial distribution different from Poisson distribution?

ANS.
Definition: The binomial distribution describes the number of successes in a fixed number of independent trials, where each trial has the same probability of success. The Poisson distribution describes the number of events that occur in a fixed interval of time or space, where the events occur randomly and independently of each other.

Number of trials: In the binomial distribution, the number of trials is fixed and known in advance. In the Poisson distribution, the number of events that will occur in a given interval is not known in advance.

Probability of success: In the binomial distribution, the probability of success is constant for each trial. In the Poisson distribution, the probability of an event occurring in a given interval is proportional to the length of the interval.

Range of values: The binomial distribution is defined for discrete integer values between 0 and the number of trials, while the Poisson distribution is defined for integer values between 0 and infinity.

Shape: The shape of the binomial distribution is symmetrical if the probability of success is 0.5. The shape of the Poisson distribution is skewed to the right.

Mean and variance: The mean and variance of the binomial distribution are both equal to np, where n is the number of trials and p is the probability of success. The mean of the Poisson distribution is equal to lambda, the parameter that describes the average number of events per interval, while the variance is also equal to lambda.

In summary, the binomial distribution is used to model the number of successes in a fixed number of trials, while the Poisson distribution is used to model the number of events that occur in a fixed interval of time or space.
